```python
# List of storing Ids and their slug
extracted_info_list = []


def extract_info_recursive_generator(data):
    if isinstance(data, list):
        for item in data:
            yield from extract_info_recursive_generator(item)
    elif isinstance(data, dict):
        yield data.get('id')

        for value in data.values():   
            yield from extract_info_recursive_generator(value)        
            
for r in res2_list:
    json2 = r.text
    json_data = json.loads(json2)
    tree = json_data.get('categoryTreeFacet', {}).get('categoryTree', [])
    t = tree[0]
    l = t.get('children',[])
    test = l[0]
    linear_t = test.get('children',[])
    extracted_info_generator = extract_info_recursive_generator(linear_t)
    extracted_info_list.extend(extracted_info_generator)


filtered_list = []
```

```python
# List to store all information from xhr requests
res3_list = []


# Sending APIs to collect core pages info
for i in filtered_list:
    url3 = f'https://de.rubix.com/rest-api/v4/search-items?pageSize=1&pageNumber=1&conditions=[{{"code":"context","values":["CATEGORY"]}},{{"code":"categoryId","values":["{i}"]}}]&facets=[]&sorts=[{{"code":"relevance"}}]&configStrategy=strategyDecompoundingReversedOrder'
    #print(url3)
    res3 = requests.get(url3, headers=headers2, data=payload)
    print(res3.status_code)
    if (res3.status_code == 200):
        res3_list.append([res3,i])
        print(url3)
    else:
        pass
```

```python
#List for collection of whole data
org_data = []

for i in res3_list:
    # Initializing response from the list
    r = i[0]
    
    # List for appending categories and supplier info
    l1 = []
    l2 = []
    l3 = []
    l4 = []
    sup_info  = {}
    sup_chunk = []
    json2 = r.text
    json_data = json.loads(json2)
    
    # Parsing Supplier Part in json response
    sp = json_data.get('facets', [])
    facet = sp[0]
    if(facet['name'] == 'Supplier'):
        sups = facet.get('values', [])
        for sup in sups:
            sup_info = {
                sup['code'] : sup['count']
            }
            sup_chunk.append(sup_info)       
    
    # Category Tree
    tree = json_data.get('categoryTreeFacet', {}).get('categoryTree', [])
    t = tree[0]
    data = t.get('children',[]) 
    
    #Function to store category heirarchy
    def store(s, lvl):       
        match lvl:
            case 1:
                l1.append(s)
            case 2:
                l2.append(s)
            case 3:
                l3.append(s)
            case 4:
                l4.append(s)
                
    level = 1
    
    # Function to traverse over Category Tree    
    def itr_children(child, level):
        for ch in child:            
            if 'name' in ch.keys():
                store(ch['name'], level)
                if 'children' in ch.keys():              
                    level += 1
                    itr_children(ch['children'], level)

    itr_children(data, level)
    link = f'https://de.rubix.com/en/linear-technology/c-{i[1]}'
    org_data.append([l1, l2, l3, l4, sup_chunk,link])
```

```python
pd.set_option('display.max_colwidth', None)
DF = pd.DataFrame(org_data, columns=['Category L1','Category L2', 'Category L3','Category L4', 'Suppliers', 'Product Page Link'])
DF
```
